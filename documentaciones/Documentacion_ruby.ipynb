# Que es Ruby:
Ruby es un lenguaje de programación de código abierto, interpretado y orientado a objetos, diseñado para ser productivo y divertido. A diferencia de PHP, Ruby no fue creado específicamente para la web, sino como un lenguaje de propósito general con una sintaxis elegante y natural.

Sin embargo, Ruby se hizo tremendamente popular para el desarrollo web gracias a Ruby on Rails, un framework que revolucionó la forma de construir aplicaciones web. Ruby se ejecuta en el servidor y, cuando se usa con Rails, procesa las peticiones, interactúa con bases de datos y genera el HTML que se envía al navegador. El usuario final solo ve el resultado HTML, nunca el código Ruby del servidor.

# 2. Operadores logicos y matematicos:

```python
%%script ruby
# Operadores Matematicos
a = 10
b = 3

suma = a + b
puts "La suma de #{a} y #{b} es: #{suma}"
resta = a - b
puts "La resta de #{a} - #{b} es: #{resta}"
multiplicacion = a * b
puts "La multiplicacion de #{a} * #{b} es : #{multiplicacion}"
=begin
 En Ruby, la división de enteros (Fixnum) resulta en un entero
Para obtener un resultado con decimales (flotante),
al menos uno de los operandos debe ser flotante.
=end
division_entera = a / b
puts "La division entera de #{a} / #{b} es: #{division_entera}"

division_flotante = a.to_f / b # Convierte 'a' a flotante para la división real
puts "La division (flotante) de #{a} / #{b} es: #{division_flotante}"
resto = a % b
puts "El resto de #{a} % #{b} es: #{resto}"
potencia = a ** b
puts "La potencia de #{a} ^ #{b} es: #{potencia}"
raiz = Math.sqrt(a)
puts "La raíz cuadrada de #{a} es: #{raiz}"


# Operadores Lógicos
x = 5
y = 10

# AND lógico (&&)
resultado_logico = (x > 0) && (y < 20)
puts "El resultado de la expresión lógica es: #{resultado_logico}"
```

# 3.Condicionales:
* IF-ELSIF-ELSE:<br>
Ruby maneja las decisiones condicionales mediante la estructura if-elsif-else, que evalúa expresiones booleanas de forma secuencial. La sintaxis utiliza if para la condición principal, elsif  para condiciones alternativas adicionales, y else como caso por defecto cuando ninguna condición anterior se cumple. Cada bloque de código asociado a una condición debe terminar con la palabra clave end, marcando el límite de la estructura condicional.
* CASE-WHEN: <br>
La estructura case-when en Ruby es la equivalente a los statements switch de otros lenguajes, pero con mayor flexibilidad y poder. Utiliza la palabra clave case seguida de la expresión a evaluar, y luego múltiples cláusulas when que especifican los valores o patrones a comparar. Una ventaja significativa de Ruby es que permite múltiples valores en una sola cláusula when separados por comas, como se muestra en when "Sábado", "Domingo".



```python
%%script ruby
edad= 17
# IF-ELSIF-ELSE
if edad < 18
    puts "Es menor de edad."
elsif edad >= 65
    puts "Es un adulto mayor."
else
    puts "Es un adulto."
end

# CASE (Equivalente al switch)
dia = "Sábado"
case dia
when "Sábado", "Domingo"
    puts "¡Es fin de semana!"
when "Lunes"
    puts "Es lunes :("
else
    puts "Es día laboral."
end
```

# 4. Bucles:
* Iteradores: <br>
Ruby maneja la iteración de manera elegante a través de bloques e iteradores, ofreciendo un enfoque más expresivo que los bucles tradicionales. El método each es el iterador fundamental que se aplica sobre rangos y colecciones, utilizando la sintaxis do...end para definir bloques de código. Para rangos numéricos como (1..3), el método each pasa cada valor al bloque a través de la variable de bloque |i|, ejecutando el código interior para cada elemento. Esta aproximación es más idiomática en Ruby que el bucle for tradicional, ya que enfatiza la iteración sobre colecciones y sigue el principio de "objetos que se comportan como colecciones". Los bloques en Ruby permiten encapsular lógica de iteración de manera clara y concisa, haciendo el código más legible y alineado con la filosofía de orientación a objetos del lenguaje.
* WHILE Y DO-WHILE:<br>
Ruby proporciona dos estructuras para bucles condicionales: while y begin...end while. El bucle while evalúa la condición al inicio de cada iteración, ejecutando el bloque de código solo si la condición es verdadera desde el principio. En contraste, la estructura begin...end while garantiza que el bloque se ejecute al menos una vez, ya que la condición se verifica al final de cada iteración. Es importante notar que Ruby utiliza += para el incremento de variables, a diferencia de los operadores ++ que no existen en el lenguaje. Ambas estructuras requieren la palabra clave do opcional para separar la condición del cuerpo del bucle cuando se escriben en una sola línea, pero pueden omitirse cuando el código se organiza en múltiples líneas con una sintaxis más explícita.
*  Iteradores de coleccion:<br>
La verdadera potencia de Ruby brilla en su manejo de colecciones a través de iteradores como each. Los arrays y otras estructuras de datos en Ruby son objetos que responden al método each, el cual itera sobre cada elemento pasándolo al bloque definido.



```python
%%script ruby
puts "Bucle FOR: "
(1..3).each do |i|
    print "#{i} "
end
puts

contador = 0
puts "Bucle WHILE: "
while contador < 3 do
    print "#{contador} "
    contador += 1
end
puts

puts "Bucle DOWHILE: "
i = 0
begin
    print "#{i} "
    i += 1
end while i < 3
puts

colores = ["Rojo", "Azul", "Amarillo"]
puts "Bucle FOREACH (Array): "
colores.each do |color|
    print "#{color} "
end
puts
```

# 5.Funciones:
En Ruby, los métodos se definen utilizando la palabra clave def seguida del nombre del método y sus parámetros entre paréntesis. Una característica distintiva de Ruby es que el return es implícito - el método automáticamente devuelve el valor de la última expresión evaluada, sin necesidad de usar la palabra clave return explícitamente.

```python
%%script ruby
def sumar(a, b)
  a + b
end
resultado = sumar(5, 3)
puts "El resultado de la suma es: #{resultado}"
```

# 6.Objetos y Encapsulamiento:
* Objetos:<br>
En Ruby, las clases se definen usando la palabra clave class y siguen el principio de que todo es un objeto. Los atributos de instancia se denotan con el prefijo @ (como @nombre y @saldo) y son privados por defecto, meaning no son accesibles directamente desde fuera de la clase. Para exponer estos atributos, Ruby proporciona los ayudantes attr_reader, attr_writer y attr_accessor que generan automáticamente los métodos getter y setter. Específicamente, attr_accessor :nombre crea tanto el getter nombre como el setter nombre= para el atributo @nombre. El método initialize funciona como el constructor de la clase y se llama automáticamente cuando se crea una nueva instancia usando Persona.new("Ricardo", 750.00). Esta aproximación de encapsulamiento asegura que el estado interno del objeto esté protegido mientras se proporciona una interface controlada para interactuar con él.
* Encapsulamiento:<br>
Ruby implementa el encapsulamiento mediante una combinación de atributos privados y métodos públicos. Los métodos de instancia como depositar(cantidad) definen el comportamiento del objeto y pueden acceder libremente a los atributos internos usando @. Para los atributos que necesitan ser leídos desde fuera de la clase, se pueden crear getters explícitos como def saldo; @saldo; end o usar los ayudantes attr_reader.A diferencia de otros lenguajes, Ruby no tiene modificadores de acceso como public o private al mismo nivel de granularidad, pero todos los métodos son públicos por defecto a menos que se especifique lo contrario, promoviendo una filosofía de "menos es más" y confiando en la disciplina del programador para mantener interfaces claras.


```python
%%script ruby
class Persona
    # attr_reader :nombre  # Crea solo un metodo getter para @nombre
    # attr_writer :nombre  # Crea solo un metodo setter para @nombre
    attr_accessor :nombre # Crea métodos getter y setter para @nombre

    # Metodo Constructor
    def initialize(nombre_inicial, saldo_inicial)
        @nombre = nombre_inicial
        @saldo = saldo_inicial
    end

    # Metodo de Comportamiento
    def depositar(cantidad)
        @saldo += cantidad
        puts "Depósito de #{cantidad} realizado."
    end

    # Metodo Getter explicito
    def saldo
        @saldo
    end
end

# Crear un Objeto (Instanciar la clase)
mi_persona = Persona.new("Luis", 20) # Llama a initialize

# Acceder a la propiedad usando el método getter (generado por attr_accessor)
puts "Nombre: #{mi_persona.nombre}"
# Acceder a la propiedad usando el método getter explícito
puts "Saldo inicial: $#{mi_persona.saldo}"

mi_persona.depositar(10)

# El método setter (generado por attr_accessor)
mi_persona.nombre = "Luis Yepez"
puts "Nuevo nombre: #{mi_persona.nombre}"

puts "Nuevo saldo: $#{mi_persona.saldo}"
```
